```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/imigrantes_canada.csv')
df
```

```python
df.info()
```

# coluna 'País'

```python
df['País']
```

```python
df.set_index('País', inplace=True)
```

```python
df
```

```python
anos = list(map(str, range(1980,2014)))
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
print(type(brasil))
```

# pandas = series e dataframes
# python = list, tuple, dict

```python
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}
```

```python
brasil_dict
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

# Gráfico de Linha

```python
! pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])

plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.grid()
plt.show()
```

# Transformando um gráfico em figura

```python
plt.figure(figsize=(10,4))
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.grid()
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.show()
```

```python
# Criando uma figura
fig,axs = plt.subplots(1,2,figsize=(15,4))


# Subplot 1: Gráfico de Linha
axs[0].plot(dados_brasil['ano'],dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title("Imigração do Brasil para o Canadá de 1980 a 2013")
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Números de imigrantes')


# Subplot 2: Gráfico Boxplot


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title("Boxplot da imigração do Brasil para o Canadá de 1980 a 2013")
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')


plt.show()
```

```python
dados_brasil.describe()
```

# Comparativo da Imigração do Brasil com alguns países da America Latina

```python
# Criar uma figura

fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

# Subplots

# Brasil
axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

# Colômbia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.set_xlabel('Ano', fontsize=10)
    ax.set_ylabel('Número de Imigrantes', fontsize=10)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))


# Usando ravel() - achatar a matriz evitando o uso complexo do for
# Usando set_ylim () - definir o limite dos valores.

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

#Aula 3 começa aqui
fig.suptitle("Imigração da América Latina para o Canadá", fontsize=18)
```

# Exercício - Verifique na documentação do Matplotlib

- função ravel() - bidimensional para 1 dimensão 
- função setylim() - set o limite do eixo y
